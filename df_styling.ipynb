```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df["species"].value_counts()
```

# Coloring species :)))

```python
def colorIt(species):
    if species == "setosa":
        color = "red"
    elif species == "virginica":
        color = "blue"
    else:
        color = "orange"
    
    return 'color: %s' % color
    
```

```python
df[['species']].style.applymap(colorIt)
```

# Styling

```python
df1 = df.style.set_table_styles(
[{"selector": "th",
  "props": [("background", "red"), 
            ("color", "white")]},

 {"selector": "tr:nth-of-type(odd)",
  "props": [("background", "#328DC9 "),
           ("color", "black")]}, 
 
 {"selector": "tr:nth-of-type(even)",
  "props": [("background", "black"),
            ("color", "white")]},
 
 {"selector": "tr:hover",
  "props": [("background-color", "green")]}]
)
```

```python
df1.caption = "Cooler df"
```
